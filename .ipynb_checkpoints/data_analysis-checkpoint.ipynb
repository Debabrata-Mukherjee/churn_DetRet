```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
pd.set_option("display.max_columns",50)
data = pd.read_csv("WA_Fn-UseC_-Telco-Customer-Churn.csv")
```

```python
data.head(5)
```

```python
churn_yes = data[data["Churn"].astype(str) == 'Yes']
churn_no = data[data["Churn"].astype(str) == 'No']
print(len(churn_yes))
print(len(churn_no))
```

```python
labels = ['Churn_YES','Churn_NO']
sizes = [len(churn_yes),len(churn_no)]
fig = plt.figure()
plt.pie(sizes,labels=labels,autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')
fig.savefig("Churn_yes_no.png")
plt.show()
```

```python
def majorPlot():
    cols = ['gender', 'SeniorCitizen', 'Partner', 'Dependents',
       'PhoneService', 'MultipleLines', 'InternetService',
       'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport',
       'StreamingTV', 'StreamingMovies', 'Contract', 'PaperlessBilling',
       'PaymentMethod']
    colors = ['lightskyblue','lightcoral']
    p = 1
    for col in cols:
        un = data[col].unique()
        for u in range(len(un)):
            plt.figure(figsize = (40,25))
            plt.subplot(21,3,p)
            ss = data[data[col].astype(str)==un[u]]
            sz = [len(ss[ss['Churn'].astype(str)=='Yes']),len(ss[ss['Churn'].astype(str)=='No'])]
            labels = ['Churn_'+str(un[u]),'No-Churn_'+str(un[u])]
            plt.pie(sizes,labels = labels,colors=colors,startangle=140)
            plt.title(str(col)+'_'+str(un[u]))
            p+=1
            if p==4:
                p=1
            plt.show()
            
```

```python
majorPlot()
```

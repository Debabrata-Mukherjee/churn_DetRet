```python
#importing the needed modules

import pandas as pd
import numpy as np
from sklearn.feature_selection import SelectFromModel
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
pd.set_option('display.max_columns',50)
data = pd.read_csv('WA_Fn-UseC_-Telco-Customer-Churn.csv')
data.head(10)
```

```python
data.shape
```

```python
data.dtypes
```

```python
data.columns
```

```python
for col in data.columns:
    print(col," : ",data[col].unique())
```

```python
#categorising and adding corresponding new_columns from existing ones
#later reducing features using feature selection techniques
def parseData(data):
    p_data = data.copy()
    cols = ['gender', 'SeniorCitizen', 'Partner', 'Dependents',
       'PhoneService', 'PaperlessBilling','Churn']
    for col in cols:
        t = list(sorted(p_data[col].unique()))
        for i in range(len(t)):
            p_data[col] = p_data[col].astype(str).replace(str(t[i]),i)
    return p_data
```

```python
p_data = parseData(data)
p_data.head(10)
```

```python
#adding categorical attributes

dum_ml = pd.get_dummies(p_data['MultipleLines'],prefix='multiple_lines')
dum_is = pd.get_dummies(p_data['InternetService'],prefix='internet_service')
dum_os = pd.get_dummies(p_data['OnlineSecurity'],prefix='online_security')
dum_ob = pd.get_dummies(p_data['OnlineBackup'],prefix='online_backup')
dum_dp = pd.get_dummies(p_data['DeviceProtection'],prefix='device_protection')
dum_ts = pd.get_dummies(p_data['TechSupport'],prefix='tech_support')
dum_stv = pd.get_dummies(p_data['StreamingTV'],prefix='streaming_tv')
dum_sm = pd.get_dummies(p_data['StreamingMovies'],prefix='streaming_movies')
dum_c = pd.get_dummies(p_data['Contract'],prefix='contract')
dum_pm = pd.get_dummies(p_data['PaymentMethod'],prefix='payment_method')
```

```python
p_data.drop(['MultipleLines', 'InternetService',
       'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport',
       'StreamingTV', 'StreamingMovies', 'Contract', 'PaymentMethod'],axis=1,inplace=True)

p_data = pd.concat([p_data,dum_ml],axis=1)
p_data = pd.concat([p_data,dum_is],axis=1)
p_data = pd.concat([p_data,dum_os],axis=1)
p_data = pd.concat([p_data,dum_ob],axis=1)
p_data = pd.concat([p_data,dum_dp],axis=1)
p_data = pd.concat([p_data,dum_ts],axis=1)
p_data = pd.concat([p_data,dum_stv],axis=1)
p_data = pd.concat([p_data,dum_sm],axis=1)
p_data = pd.concat([p_data,dum_c],axis=1)
p_data = pd.concat([p_data,dum_pm],axis=1)
p_data.head(5)
```

```python
#converting the data to numeric form for calculations

for col in p_data.columns:
    if col!='customerID':
        p_data[col] = pd.to_numeric(p_data[col],errors="coerce")
p_data.dtypes
```

```python
#no need of customerID in training and testing
modelData = p_data.loc[:,p_data.columns!='customerID']
modelData[modelData==np.inf]=np.nan
modelData.fillna(modelData.mean(),inplace=True)
```

```python
x_orig = modelData.loc[:,modelData.columns!="Churn"]
y_orig = modelData["Churn"]

x_orig,y_orig = np.asarray(x_orig),np.asarray(y_orig)
print(x_orig.shape)
print(y_orig.shape)
```

```python
x_orig = preprocessing.StandardScaler().fit(x_orig).transform(x_orig)
```

```python
#splitting the data into 80% training and 20% testing
X_train,X_test,y_train,y_test = train_test_split(x_orig,y_orig,test_size=20/100,random_state=60)
print(X_train.shape)
print(X_test.shape)
```

```python
#selecting important features using random forest classifier

select = SelectFromModel(RandomForestClassifier(n_estimators=100,random_state=37),threshold='median')
select.fit(X_train,y_train)
X_train_s = select.transform(X_train)
print("Shape of X_train: ",X_train.shape)
print("Shape of X_train_s: ",X_train_s.shape)
```

```python
mask = select.get_support()
plt.matshow(mask.reshape(1,-1),cmap='gray_r')
plt.xlabel('Index of features (black ones are the selected features)')
```

```python
X_test_s = select.transform(X_test)

lr = LogisticRegression()
score = lr.fit(X_train_s,y_train).score(X_test_s,y_test)
print("Model Accuracy = {:.3f}%".format(score*100))
```
